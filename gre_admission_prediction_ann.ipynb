```python
import pandas as pd

df = pd.read_csv('/content/Admission_Predict.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.drop('Chance of Admit ',axis=1),df['Chance of Admit '],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train = x_train.drop('Serial No.',axis = 1)
x_test = x_test.drop('Serial No.',axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers,Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(7,activation='relu',input_dim = 7),
    Dense(10,activation='relu'),
    Dense(1,activation = 'linear')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
history = model.fit(x_train,y_train,epochs = 100,batch_size = 32,validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)*100
```
